```python
import pandas as pd

txt_folder = 'txt_files'
csv_folder = 'csv_files'
sep = '\^B'
```

```python
df_chunk = pd.read_csv(f'{txt_folder}/TB_KT15.txt', encoding='utf-8', sep=sep, chunksize=100000)
```

```python
cnt = 1
p_1 = 1
p_2 = 1
merge_19 = pd.DataFrame()
merge_20 = pd.DataFrame()

for df in df_chunk:

    df_19 =  df[df['출원번호'] // 10000000 - 400000 == 2019]
    df_20 =  df[df['출원번호'] // 10000000 - 400000 == 2020]

    df_19 = df_19.fillna('').groupby(['출원번호', '지정상품한글명', '류'])['유사군'].apply(' '.join).reset_index()
    df_20 = df_20.fillna('').groupby(['출원번호', '지정상품한글명', '류'])['유사군'].apply(' '.join).reset_index()
    
    l_19 = len(df_19)
    l_20 = len(df_20)
    
    if l_19:
        merge_19 = pd.concat([merge_19, df_19])
    if l_20:
        merge_20 = pd.concat([merge_20, df_20])
        
    if cnt % 10 == 0:
        if len(merge_19):
            merge_19['유사군'] = merge_19['유사군'].apply(lambda x: sorted(x.split(' '), reverse=False))
            merge_19['유사군'] = merge_19['유사군'].apply(' '.join)
            merge_19.to_csv(f'{csv_folder}/19년도_{p_1}.csv', index=False, encoding='utf-8')
            merge_19 = pd.DataFrame()
            p_1 += 1
            
        if len(merge_20):
            merge_20['유사군'] = merge_20['유사군'].apply(lambda x: sorted(x.split(' '), reverse=False))
            merge_20['유사군'] = merge_20['유사군'].apply(' '.join)
            merge_20.to_csv(f'{csv_folder}/20년도_{p_2}.csv', index=False, encoding='utf-8')
            merge_20 = pd.DataFrame()
            p_2 += 1
        
    print(cnt, l_19, l_20)
    cnt += 1

merge_19.to_csv(f'{csv_folder}/19년도_last.csv', index=False, encoding='utf-8')
merge_20.to_csv(f'{csv_folder}/20년도_last.csv', index=False, encoding='utf-8')
```
