```python
import pandas as pd
import os
txt_folder = 'txt_files'
csv_folder = 'csv_files'
sep = '\^B'

df_applicant_19 = pd.read_csv(f'{csv_folder}/applicant_19.csv', encoding='utf-8')
df_applicant_20 = pd.read_csv(f'{csv_folder}/applicant_20.csv', encoding='utf-8')

df_applicant = pd.concat([df_applicant_19, df_applicant_20])

df_agent_19 = pd.read_csv(f'{csv_folder}/agent_19.csv', encoding='utf-8')
df_agent_20 = pd.read_csv(f'{csv_folder}/agent_20.csv', encoding='utf-8')

df_agent = pd.concat([df_agent_19, df_agent_20])

# df_applicant.set_index('출원번호', inplace=True)
# df_agent.set_index('출원번호', inplace=True)
```

```python
df_applicant['출원인코드'] = df_applicant['출원인코드'].map(str)

df_agent['대리인코드'] = df_agent['대리인코드'].map(str)
```

```python
df_applicant_merged = df_applicant.groupby('출원번호').agg({
    '출원인코드': ' '.join
})
df_agent_merged = df_agent.groupby('출원번호').agg({
    '대리인코드': ' '.join
})
# del df_temp_applicant
display(df_applicant_merged.head(10))
display(df_agent_merged.head(10))
```

```python
df_19_merged = pd.read_csv(f'{csv_folder}/19년도_병합.csv', encoding='utf-8', low_memory=False)
df_20_merged = pd.read_csv(f'{csv_folder}/20년도_병합.csv', encoding='utf-8', low_memory=False)

display(df_19_merged.head())
```

```python
df_19_merged['출원인코드'] = df_19_merged['출원번호'].apply(lambda x: df_applicant_merged.loc[x]['출원인코드'] if x in df_applicant_merged.index else 0)
df_19_merged['대리인코드'] = df_19_merged['출원번호'].apply(lambda x: df_agent_merged.loc[x]['대리인코드'] if x in df_agent_merged.index else 0)

df_19_merged.to_csv(f'{csv_folder}/19년도_병합_출원인대리인코드.csv', index=False, encoding='utf-8')
```

```python
df_20_merged['출원인코드'] = df_20_merged['출원번호'].apply(lambda x: df_applicant_merged.loc[x]['출원인코드'] if x in df_applicant_merged.index else 0)
df_20_merged['대리인코드'] = df_20_merged['출원번호'].apply(lambda x: df_agent_merged.loc[x]['대리인코드'] if x in df_agent_merged.index else 0)

df_20_merged.to_csv(f'{csv_folder}/20년도_병합_출원인대리인코드.csv', index=False, encoding='utf-8')
```
